```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns 
import plotly.graph_objects as go
```

```python
filename = 'Votes 2019.xls'
df = pd.read_excel(filename)
```

```python
print(df.head())
```

```python
totalelectors = df['Total Electors'].sum()
print(totalelectors)
```

```python
actualvoters = df['Total Actual Votes'].sum()
print(actualvoters)
```

```python
femaletotal = df['Female '].sum()
maletotal = df['Male'].sum()
print(maletotal, femaletotal)
```

```python
diff = list(df['Female ']>df['Male'])
```

```python
votersturnout = (actualvoters/totalelectors)*100
```

```python
df['State Name'] = df['State Name'].str.replace('*','', regex=True)
```

```python
df['Voters Turnout Percentage'] = df['Voters Turnout Percentage'].round(1)
```

```python
%matplotlib notebook
#sns.set_style('darkgrid')
width=.75
fig = go.Figure()
plt.figure(figsize=(15, 20))  
colors= ['blue',] * 36

index=0
for i in diff:
    if i:
        colors[index]='crimson'
    index+=1

plt.barh(df['State Name'], df['Voters Turnout Percentage'], width, color=colors)
plt.yticks(np.arange(36), df['State Name'])

for i, v in enumerate(df['Voters Turnout Percentage']):
    plt.text(v+1 , i , str(v), color='black')

# fig.add_layout_image(
#         go.layout.Image(
#             source="https://images.plot.ly/language-icons/api-home/python-logo.png",
#             xref="x",
#             yref="y",
#             x=0,
#             y=20,
#             sizex=15,
#             sizey=20,
#             sizing="stretch",
#             opacity=0.5,
#             layer="below")
# )
plt.grid(color='blue', linestyle='-', linewidth=0.25, alpha=.25)
plt.xlabel('Percentage')
plt.ylabel('State')
plt.title('Voters Turnout Percentage Statewise')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.axis([0, 100, None, None])
plt.text(80, 37, 'Blue with more male voters',bbox=dict(facecolor='blue', alpha=0.5))
plt.text(80, 36, 'Red with more female voters', bbox=dict(facecolor='crimson', alpha=0.5))
plt.show()
```
