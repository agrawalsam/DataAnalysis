```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
filename = 'Donations to national political parties - India.xlsx'
df = pd.read_excel(filename)
```

```python
del df['Unnamed: 8']
del df['Unnamed: 9']
del df['Unnamed: 11']
del df['Unnamed: 12']
del df['Unnamed: 10']
del df['Unnamed: 13']
```

```python
df = df.iloc[:,[0,5,7]]
```

```python
%matplotlib notebook
df.groupby(['Party','Financial Year '])['Amount'].sum().unstack('Party').plot.bar()
plt.axis([None,None,0,4500000000])
plt.title('Donations Received By Different Parties')
plt.ylabel('Donations Received')
plt.legend()
plt.show()
```
