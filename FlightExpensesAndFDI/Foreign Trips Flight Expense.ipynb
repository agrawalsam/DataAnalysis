```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
filename = 'PM Modi Foreign Visits.xls'
df = pd.read_excel(filename)
```

```python
df = df.iloc[:,[1,4,5]]
print(df.head())
df['Expenses incurred on Chartered Flight (Rs.)'] = [0 if(x=='IAF BBJ Aircraft' or x=='Bill not received') 
                                                     else x for x in df['Expenses incurred on Chartered Flight (Rs.)']]
print(df)
df = df[df['Expenses incurred on Chartered Flight (Rs.)']>0.0]
print(df)
```

```python
df1 = df.groupby('Place of Visit').sum()
```

```python
x9 = df1.index.get_level_values(0).tolist()
print(x9)
```

```python
%matplotlib notebook
df1['No of Days'].plot.bar(color='b', align='center')
plt.xlabel('Countries')
plt.ylabel('No. Of Days')
plt.xticks([r for r in range(57)],x9, rotation=90)
plt.title('No. of Days Spent in Countries')
plt.show()
```

```python
plt.savefig('Abroad.png')
```

```python
%matplotlib notebook
df1['Expenses incurred on Chartered Flight (Rs.)'].plot.bar(color='g', align='center', 
                                                            edgecolor='white')
plt.xlabel('Countries')
plt.ylabel('Expenses incurred on Chartered Flight (Rs.)')
plt.xticks([r for r in range(57)],x9, rotation=90)
plt.title('Expenses of flights while visiting countries by PM Modi')
plt.ticklabel_format(style='plain', axis='y')
plt.axis([None, None, 0, 500000000])
plt.show()
```

```python
plt.savefig('FlightsExpenditure.png')
```
