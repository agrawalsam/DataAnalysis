```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_excel('IndianStudentsAbroad.xls')
```

```python
df1 = df.iloc[:86,[1,2,3]]
```

```python
print(df1)
```

```python
df2 = df1.sort_values(by='Percentage', ascending=False)
df2['Country'] = df2['Country'].replace('\n','',regex=True)
print(df2)
```

```python
dftop = df2.iloc[0:11,[0,2]]
print(dftop)
```

```python
dftop = dftop.iloc[0:7,:]
dftop = dftop.append([{'Country':'Others','Percentage':16.88493}])
print(dftop)
```

```python
%matplotlib notebook
explode = (0,0.2,0,0.2,0,0.2,0,0.2)
colors= ('#adb0ff', '#ffb3ff', '#90d595', '#e48381','#aafbff', '#f7bb5f', '#eafb50','#00FFFF')
plt.pie(dftop['Percentage'], labels=dftop['Country'], explode=explode, colors=colors, autopct='%1.1f%%')
plt.title('Indian Students studying in Other Countries')
plt.show()
```

```python
plt.savefig('%agecountry.png')
```
