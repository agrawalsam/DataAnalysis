```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_excel('IndianStudentsAbroad.xls')
```

```python
ldf = pd.read_excel('IndianStudentsAbroadLatest.xls')
```

```python
df1 = df.iloc[:86,[1,2,3]]
```

```python
ldf1 = ldf.iloc[:90,[0,1,2]]
```

```python
print(df1)
```

```python
print(ldf1)
```

```python
df2 = df1.sort_values(by='Percentage', ascending=False)
df2['Country'] = df2['Country'].replace('\n','',regex=True)
print(df2)
```

```python
ldf1 = ldf1.sort_values(by='Percentage', ascending=False)
ldf1['Country'] = ldf1['Country'].replace('\n','',regex=True)
print(ldf1)
```

```python
dftop = df2.iloc[0:11,[0,2]]
print(dftop)
```

```python
ldftop = ldf1.iloc[0:11,[0,2]]
print(ldftop)
```

```python
dftop = dftop.iloc[0:7,:]
dftop = dftop.append([{'Country':'Others','Percentage':16.88493}])
print(dftop)
```

```python
ldftop = ldftop.iloc[0:7,:]
ldftop = ldftop.append([{'Country':'Others','Percentage':20.180942}])
print(ldftop)
```

```python
%matplotlib notebook
explode = (0.1,0.1,0.1,0.3,0.5,0.7,0.9,0.5)
colors= ('#adb0ff', '#ffb3ff', '#90d595', '#e48381','#aafbff', '#f7bb5f', '#eafb50','#00FFFF')
plt.subplot(1,2,1)
plt.pie(dftop['Percentage'], labels=dftop['Country'], explode=explode, colors=colors, autopct='%1.1f%%')
plt.title('As on 8th Aug, 2017')

plt.subplot(1,2,2)
plt.pie(ldftop['Percentage'], labels=ldftop['Country'], explode=explode, colors=colors, autopct='%1.1f%%')
plt.title('As on 18th July, 2018')
plt.tight_layout()

plt.suptitle('Indian Students Studying Abroad')
plt.show()
```

```python
plt.savefig('%agecountry.png')
```
